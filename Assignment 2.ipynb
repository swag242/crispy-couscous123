```python
#c
df=df.drop(['CustomerID', 'AddressLine1', 'PhoneNumber', 'BirthDate'], axis=1)
df.head()
```

```python
#c
df['City']=pd.factorize(df['City'])[0]
df['StateProvinceName']=pd.factorize(df['StateProvinceName'])[0]
df['CountryRegionName']=pd.factorize(df['CountryRegionName'])[0]
df['Gender'] = pd.factorize(df['Gender'])[0]
df['Occupation'] = pd.factorize(df['Occupation'])[0]
df['Gender'] = pd.factorize(df['Gender'])[0]
df['MaritalStatus'] = pd.factorize(df['MaritalStatus'])[0]
df['PostalCode'] = pd.factorize(df['PostalCode'])[0]

df
```

```python
#c
import numpy as np
bins=np.linspace(min(df['YearlyIncome']), max(df['YearlyIncome']), 3)
group_names=['Rich', 'Middle Class', 'Poor']
pd.cut(df['YearlyIncome'], bins, group_names, include_lowest=True)
```

```python
#d
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df['YearlyIncome']=scaler.fit_transform(df['YearlyIncome'].values.reshape(-1, 1)).flatten()
df['NumberCarsOwned']=scaler.fit_transform(df['NumberCarsOwned'].values.reshape(-1, 1)).flatten()
```

```python
#e
df['PostalCode'].value_counts()
```

```python
#e
from sklearn.preprocessing import OneHotEncoder

df1=df[['Gender', 'MaritalStatus']].copy()

df=df.drop(['Gender', 'MaritalStatus'], axis=1)
df=df.join(df1)

enc=OneHotEncoder()
enc_data=pd.DataFrame(enc.fit_transform(df1[['Gender', 'MaritalStatus']]).toarray())
  
df=df.join(enc_data)

df.head()
```
